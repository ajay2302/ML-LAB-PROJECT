```python
import pandas as pd #import dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dataset 
file_path = "Full_Model_Output.xlsx"
data = pd.read_excel(file_path)
data
```

```python
#A1
```

```python
from sklearn.linear_model import LinearRegression

X_train = data[['we']].to_numpy()
y_train = data['True_Label'].replace('Non-Fact',0).replace('Fact',1).to_numpy()
reg = LinearRegression().fit(X_train,y_train)
y_predicted = reg.predict(X_train)
print(y_predicted)
```

```python
#A2
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

def get_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse) # RMSE is the square root of MSE
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return [mse, rmse, mape, r2]

train_results = get_metrics(y_train, y_predicted)
METRICS = ['mse','rmse','mape','r2']
for i,j in zip(METRICS,train_results):
    print(f'{i} : {j}')
```

```python
#A3
```

```python
X_train = data[['we','were','what','when']].to_numpy()
y_train = data['True_Label'].replace('Non-Fact',0).replace('Fact',1).to_numpy()
reg = LinearRegression().fit(X_train,y_train)
y_predicted = reg.predict(X_train)

results = get_metrics(y_train, y_predicted)
METRICS = ['mse','rmse','mape','r2']
for i,j in zip(METRICS,results):
    print(f'{i} : {j}')
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0,
n_init="auto").fit(X_train)
kmeans.labels_,kmeans.cluster_centers_
```

```python
#A5
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
kmeans = KMeans(n_clusters=2, random_state=42).fit(X_train)
silhouette_score(X_train, kmeans.labels_)
calinski_harabasz_score(X_train, kmeans.labels_)
davies_bouldin_score(X_train, kmeans.labels_)
```

```python
k = [2,3,4,5,7,9,11,8,6,7]
ans = []
for i in k:
    kmeans = KMeans(n_clusters=i, random_state=0,
    n_init="auto").fit(X_train)
    ans.append(len(kmeans.cluster_centers_))

plt.plot(ans,k)
```
